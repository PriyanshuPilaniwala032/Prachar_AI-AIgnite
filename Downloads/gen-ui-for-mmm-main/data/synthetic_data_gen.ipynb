```python
import pandas as pd
from sdv.metadata import SingleTableMetadata
from sdv.single_table import GaussianCopulaSynthesizer
```

```python
import pandas as pd

data = pd.read_csv('/Users/mike/GitHub/gen-ui-for-mmm/synthetic_data/home_equity_loan_synth_data_seed.csv')



# create metadata from schema:

# CREATE TABLE loan_applications (
#     application_id VARCHAR(15) PRIMARY KEY,
#     borrower_id VARCHAR(10) NOT NULL,
#     application_date DATE NOT NULL,
#     approval_date DATE,
#     time_to_decision INTEGER,
#     application_status VARCHAR(10) NOT NULL CHECK (application_status IN ('Approved', 'Denied', 'Pending', 'Withdrawn')),
#     amount_requested DECIMAL(10, 2) NOT NULL,
#     approved_amount DECIMAL(10, 2),
#     interest_rate DECIMAL(4, 2),
#     term INTEGER CHECK (term IN (120, 180, 240, 360)),
#     ltv_ratio DECIMAL(5, 2) NOT NULL,
#     combined_ltv DECIMAL(5, 2) NOT NULL,
#     loan_purpose VARCHAR(20) NOT NULL,
#     credit_score INTEGER NOT NULL,
#     annual_income DECIMAL(10, 2) NOT NULL,
#     property_value DECIMAL(12, 2) NOT NULL
# );

metadata = SingleTableMetadata()
metadata.add_column('application_id', sdtype='id', pii=False)
metadata.add_column('borrower_id', sdtype='id', pii=False)
metadata.add_column('application_date', sdtype='datetime', pii=False)
metadata.add_column('approval_date', sdtype='datetime', pii=False)
metadata.add_column('time_to_decision', sdtype='numerical', pii=False)
metadata.add_column('application_status', sdtype='categorical', pii=False)
metadata.add_column('amount_requested', sdtype='numerical', pii=False)
metadata.add_column('approved_amount', sdtype='numerical', pii=False)
metadata.add_column('interest_rate', sdtype='numerical', pii=False)
metadata.add_column('term', sdtype='categorical', pii=False)
metadata.add_column('ltv_ratio', sdtype='numerical', pii=False)
metadata.add_column('combined_ltv', sdtype='numerical', pii=False)
metadata.add_column('loan_purpose', sdtype='categorical', pii=False)
metadata.add_column('credit_score', sdtype='numerical', pii=False)
metadata.add_column('annual_income', sdtype='numerical', pii=False)
metadata.add_column('property_value', sdtype='numerical', pii=False)


synthesizer = GaussianCopulaSynthesizer(metadata)
synthesizer.fit(data)
synthetic_data = synthesizer.sample(num_rows=10000)
synthetic_data.to_csv('synthetic_data_gen.csv', index=False)
```
